```python
a = int(input("Enter the number you want to check palindrome :"))
b = a
rev = 0
while a > 0 :
  digit = a % 10
  rev = rev * 10 + digit
  a //= 10

print(rev)
if b == rev :
  print("Number is palindrome ")
else :
  print("Number is not palindrome ")
```

```python
# 5 : Print a pyramid pattern like:
#    *
#   ***
#  *****
# *******

for i in range(1, 6):
    print(" " * (5 - i) + "*" * (2 * i - 1))

for i in range(1, 6):
    for j in range(1, 6 - i):
        print(" ", end="")
    for j in range(1, 2 * i):
        print("*", end="")
    print()
```
