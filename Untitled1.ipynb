```python
numbers = [10, -5, 20, -10, 30, -2]
def sum(nums):
    pos = 0
    for num in nums:
        if num > 0:
            pos+= num
    return pos
result = sum(numbers)
print(result)
```

```python
sentence = input("Enter a sentence: ")
vowels = "aeiou"
count = 0
for char in sentence.lower():
    if char in vowels:
        count += 1
print("Total number of vowels:", count)
```
